```python
import numpy as np 
import pandas as pd 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import requests
import requests 
print('Libraries imported.')
```

```python

c_wikiPage = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
c_data=c_wikiPage.text
x=c_data.find("<table>")
y=c_data.find("</table>")
c_data=c_data[x+6:y]
x=c_data.find("<th>")
y=c_data.find("</th>")
col1=c_data[x+4:y]
c_data=c_data[y+5:]
x=c_data.find("<th>")
y=c_data.find("</th>")
col2=c_data[x+4:y]
c_data=c_data[y+4:]
x=c_data.find("<th>")
y=c_data.find("</th>")
col3=c_data[x+4:y-1]
c_data=c_data[y+2:]
columns_Data=[col1,col2,col3]
CanData=c_data
columns_Data
```

```python
c_data=CanData
CanList=[]
tdExists=True
while tdExists:
    i=0
    Item=[]
    while i<3:
        x=c_data.find("<td>")
        y=c_data.find("</td>")
        if x==-1:
            tdExists=False
        v=c_data[x+4:y]
        if v.find("title") != -1:
            v=v[v.find("title=\""):]
            m=v.find(">")
            n=v.find("</a>")
            v=v[m+1:n]
        v=v.replace("\n","")
        Item.append(v)
        c_data=c_data[y+5:]
        i+=1
    if 'Not assigned' not in str(Item):
        CanList.append(Item)
del CanList[-1]

CanDat=pd.DataFrame(CanList, columns=columns_Data)
CanDat.shape#data shape
CanDat
```
